```
client = Client()
spreads = client.find_spreadsheet_files_in_folders('Dokumenttien kuvailut')

df_kuvailut1 = pd.DataFrame()
df_kuvailut2 = pd.DataFrame()
df_kuvailut3 = pd.DataFrame()
df_kuvailut4 = pd.DataFrame()

kuvailu_id = 0
cols_to_keep = ['id', 'nimeke', 'url kokotekstiin', 'aiheet', 'kid']

for spread in spreads['Dokumenttien kuvailut']:
  print(spread)
  # Dokumentit viime työpajasta:
  if spread['name'].startswith("Copy"):
      df_sheet = Spread(spread['id'], sheet=0).sheet_to_df(index=None)
      df_sheet['kid'] = kuvailu_id
      df_sheet.columns = df_sheet.columns.str.lower()
      df_kuvailut1 = df_kuvailut1.append(df_sheet[cols_to_keep])
  # Satakunnan kansan artikkelit:
  elif spread['name'].startswith("SKT"):
      df_sheet = Spread(spread['id'], sheet=0).sheet_to_df(index=None)
      df_sheet['kid'] = kuvailu_id
      df_sheet.columns = df_sheet.columns.str.lower()
      df_kuvailut2 = df_kuvailut2.append(df_sheet[cols_to_keep])    
  # JYX opinnäytteet:
  elif spread['name'].startswith("jyx"):
      df_sheet = Spread(spread['id'], sheet=0).sheet_to_df(index=None)
      df_sheet['kid'] = kuvailu_id
      df_sheet.columns = df_sheet.columns.str.lower()
      df_kuvailut3 = df_kuvailut3.append(df_sheet[cols_to_keep])    
  # Tietokirjat:
  elif spread['name'].startswith("kuvailut_tieto"):
      df_sheet = Spread(spread['id'], sheet=0).sheet_to_df(index=None)
      df_sheet['kid'] = kuvailu_id
      df_sheet.rename(columns={'isbn': 'url kokotekstiin'}, inplace=True)
      df_kuvailut4 = df_kuvailut4.append(df_sheet[cols_to_keep])
  kuvailu_id += 1
  if kuvailu_id > 10: 
    break
df_kuvailut = pd.concat([df_kuvailut1, df_kuvailut2, df_kuvailut3, df_kuvailut4])
df_kuvailut.rename(columns={'id': 'did'}, inplace=True)
```
